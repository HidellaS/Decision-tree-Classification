# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the dataset

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

```python
iris['target']
```

```python
iris['target_names']
```

# Seperate into independent and dependent features

```python
X = pd.DataFrame(iris['data'], columns=['sepal length in cm', 'sepal width in cm', 'petal length in cm', 'petal width in cm'])
X
```

```python
y = iris['target']
y
```

# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1011)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

***********

#### In decision tree we dont need to do any kind of scaling (feature scaling like standadization). we can directly apply the algorithm

# Create the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier.fit(X_train, y_train)
```

# Visualize the decision tree

```python
from sklearn import tree
```

```python
plt.figure(figsize=(10, 7))
tree.plot_tree(treeclassifier, filled=True);
plt.tight_layout()
```

# Check the performance

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
y_pred =treeclassifier.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
creport = classification_report(y_test, y_pred)
print(creport)
```

```python
ac = accuracy_score(y_test, y_pred)
ac
```

# Decision treee with max depth=2

```python
treeclassifier2 = DecisionTreeClassifier(max_depth=2)
treeclassifier2.fit(X_train, y_train)
```

```python
plt.figure(figsize=(10, 7))
tree.plot_tree(treeclassifier2, filled=True);
plt.tight_layout()
```

# Check the performance

```python
y_pred2 =treeclassifier2.predict(X_test)
```

```python
cm2 = confusion_matrix(y_test, y_pred2)
cm2
```

```python
creport2 = classification_report(y_test, y_pred2)
print(creport2)
```

```python
ac2 = accuracy_score(y_test, y_pred2)
ac
```

# As we can see by going to maxdepth 2 we only miss classify 1 point so this is ok

# Prepruning and hyper parameter tuning

```python
params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1, 2, 3, 4, 5, 6],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
params
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
treemodel = DecisionTreeClassifier()
```

```python
grid = GridSearchCV(estimator=treemodel, param_grid=params, cv=5, scoring='accuracy')
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
grid.best_score_
```

# Check the performance of grid model

```python
y_pred_grid = grid.predict(X_test)
y_pred_grid
```

```python
cmg = confusion_matrix(y_test, y_pred_grid)
cmg
```

```python
creportg = classification_report(y_test, y_pred_grid)
print(creportg)
```

```python
acg = accuracy_score(y_test, y_pred_grid)
acg
```
